```
import tensorflow as tf 
from tensorflow import keras
import numpy as np
import re as re
```

```
righe_file = open('romeo/film0.txt',encoding='utf-8',errors='ignore').read().split('\n')

#cleaning text rows
cleaned_righe_file=[]
for sentence in righe_file:
  sentence = re.sub(r"([\-])","",sentence)
  sentence = re.sub(r"([\''])","",sentence)
  sentence = re.sub(r"([0-9])","",sentence)
  cleaned_righe_file.append(sentence)
#print(cleaned_righe_file)
```

```
#create dictionary
wordsCountMap={}

for sentence in cleaned_righe_file:
  splitted = sentence.split()
  for word in splitted:
    if word not in wordsCountMap:
      wordsCountMap[word]=1
    else:
      wordsCountMap[word]+=1

#create a list of word
dictionary=[]
for wordx in wordsCountMap:
  dictionary.append(wordx)
  
dictionary.sort()

#from word to int index
words_to_index_map={}
for i in range(len(dictionary)):
  words_to_index_map[dictionary[i]]=i

#for i in range(len(dictionary)):
#  print(words_to_index_map[dictionary[i]]," ",dictionary[i])

MAX_WORD_PER_SENT=50
```

```
#create string input array
inputrows=[]
counter = 0
for sentence in cleaned_righe_file:
  row=[]
  splitted = sentence.split()
  for wIndex in range(MAX_WORD_PER_SENT):
    if wIndex<len(splitted):
      row.append( words_to_index_map[splitted[wIndex]]/len(dictionary) )
    else:
      row.append( 0 )
  counter+=1
  inputrows.append(row)
  
  
print(inputrows)


```

```
labels=[]
features=[]
for i in range(0,len(inputrows)-1 ):
  features.append(inputrows[i])
  ix = []
  
  for k in range(0,len(inputrows)  ):
    ix.append(0.0)
    
  ix[i+1] = 1
  labels.append(ix)

print(len(labels[0])," ",len(features[0])," ",len(inputrows)," ",len(features)," ",len(labels) )
```

```
print( (len(inputrows)) )
model = keras.Sequential()
model.add(keras.layers.Dense(512,activation="relu",input_dim=MAX_WORD_PER_SENT))
model.add(keras.layers.Dense(512,activation="relu"))
model.add(keras.layers.Dense(671,activation="softmax"))
```

```
# Configure a model for categorical classification.
model.compile(optimizer=tf.train.RMSPropOptimizer(0.01),
              loss=tf.keras.losses.categorical_crossentropy,
              metrics=[tf.keras.metrics.categorical_accuracy])
```

```
#fit the model
dataIn = np.array(features)
dataLbs = np.array(labels)
#dataLbs = np.zeros( (670,671) )

#dataIn = np.random.random((670, 50))
#dataLbs = np.random.random((1000, 671))

model.fit(dataIn, dataLbs, epochs=1000, batch_size=128)
print(len(labels)," ",len(features))
```
